```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import binom

%matplotlib inline
```

### Equations for number 1

```python
#equations 
f_x= lambda x:1/(1-x)
f_x1=lambda x: 1/(1-x)**2
f_x2=lambda x: -2/(x-1)**3

#theoretical values 
x=0.1
dx=0.05
```

### Function diff_cen

```python
def diff_cen(f,x,dx,deg):
    if deg == 1:
        d1=(f_x(x+dx)-f_x(x))/dx
        print(f'f\'({x})= {d1}, error @ {abs(f(x)-d1)}')
    elif deg == 2:
        d2=(f_x(x+2*dx)-2*f_x(x+dx)+f_x(x))/dx**2
        print(f'f\'({x})= {d2}, error @ {abs(f(x)-d2)}')
    elif deg == 3:
        d3=(f_x(x+3*dx)-3*f_x(x+2*dx)+3*f_x(x+dx)-f_x(x))/(dx**3)
        print(f'f\'({x})= {d3}, error @ {abs(f(x)-d3)}')
    elif deg <= 4:
          print('the desired degree is not covered by the function')
    
```

### Output

```python
diff_cen(f_x,0.1,0.05,1)
diff_cen(f_x1,0.1,0.05,2)
diff_cen(f_x2,0.1,0.05,3)
diff_cen(f_x,0.1,0.05,4)
```

### Equations for number 2

```python
y1= lambda x:((4*x**2+2*x+1)/(x+2*e**x))**x
y2= lambda x: np.cos(2*x)+(x**2/20)+e**-2*x
e=np.log(1+x)
```

```python
binom_coeff=lambda n: np.asarray([binom(n,k) for k in range(n+1)])# this line will be used for the pascal triangle

def diff_fwd(f,x,dx,deg):
    f_ans=f(x+((np.arange(deg,-1,-1))*dx))
    bin= binom_coeff(deg)
    bin[1::2]*= -1
    diff=(bin @ f_ans) / dx**deg
    return print(f'f\'({x})= {diff}, error @ {abs(f(x)-diff)}')
```

```python
def diff_bwd(f,x,dx,deg):
    f_ans=f(x+((np.arange(0,-(deg+1),-1))*dx))
    bin= binom_coeff(deg)
    bin[1::2]*= -1
    diff=(bin @ f_ans) / dx**deg
    return print(f'f\'({x})= {diff}, error @ {abs(f(x)-diff)}')
```

```python
diff_fwd(y1,0.1,0.05,1)
```

```python
diff_bwd(y2,0.1,0.05,1)
```

 ### Taylor Series expansion 

```python
X= np.linspace(0,.5,10) # array of x values X < 1

#equation 
f= 4*X**2+2*X+1/(X+2*e**X)
f_2= np.cos(2*X)+(X**2/20)+(e**-2*X)
f_0= 1/(1-X)
```

complete taylor expansion of this function is:<br>
$$f(x)=1+x+x^2+x^3+...$$<br>
0th order approximation is:<br>
$$f(x)=1$$<br>
1st order approximation is:<br>
$$f(x)=1+X$$<br>
2nd order approximation is:<br>
$$f(x)=1+X+X**2$$

```python
plt.figure()
plt.plot(X,f,'k',label='$f=1/(1-x)$')
plt.title('plot for equation  $y1=4X^2+2X+1/(X+e^-2x)$')
plt.grid()
plt.legend()
```

```python
f0=1+np.zeros(np.size(X))
f1=1+X
f2=1+X+X**2
plt.figure()
plt.plot(X,f0,'b',label='$f=1$')#0th order approximation
plt.plot(X,f1,'g',label='$f=1+X$')#1st order approximation
plt.plot(X,f2,'r',label='$f=1+X+X**2$')#2nd order approximation
plt.errorbar(X,f,yerr=0.05,color='k',label='$y1=4X^2+2X+1/(X+ e^-2x)$')#equation
plt.title(' power series approximating for y1')
plt.grid()
plt.legend()
```

```python
plt.figure()
plt.plot(X,f_2,'k',label='$y2=cos(2X)+(X^2/20)+e^2x$')
plt.title('plot for equation  $y2=cos(2X)+(X^2/20)+e^2x$')
plt.grid()
plt.legend()
```

```python
f0=1+np.zeros(np.size(X))
f1=1+X
f2=1+X+X**2
plt.figure()
plt.plot(X,f0,'b',label='$f=1$')#0th order approximation
plt.plot(X,f1,'g',label='$f=1+X$')#1st order approximation
plt.plot(X,f2,'r',label='$f=X+X**2$')#2nd order approximation
plt.errorbar(X,f_2,yerr=0.05,color='k',label='$y2=cos(2X)+(X^2/20)+e^2x$')#equation
plt.title(' power series approximating for y2')
plt.grid()
plt.legend()
```

```python
plt.figure()
plt.plot(X,f_0,'k',label='$f=1/(1-x)$')
plt.title('plot for equation  $y2=cos(2X)+(X^2/20)+e^2x$')
plt.grid()
plt.legend()
```

```python
f0=1+np.zeros(np.size(X))
f1=1+X
f2=1+X+X**2
plt.figure()
plt.plot(X,f0,'b',label='$f=1$')#0th order approximation
plt.plot(X,f1,'g',label='$f=1+X$')#1st order approximation
plt.plot(X,f2,'r',label='$f=1+X+X**2$')#2nd order approximation
plt.errorbar(X,f_0,yerr=0.05,color='k',label='$f=1/(1-x)$')#equation
plt.title(' power series approximating for sample number 1')
plt.grid()
plt.legend()
```

the vertical bars indicates the precesision with which we justified in estimating the value of f
